```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# Data Creation
df=pd.DataFrame({'Circles':['Ambala','Panchkula','Chandigarh','Mohali','Zirakpur','Kharar','Derabassi','Rajpura','Patiala','Ludhiana'],
                 'Division':['Ambala','Ambala','Chandigarh','Chandigarh','Chandigarh','Chandigarh','Chandigarh','Patiala','Patiala','Ludhiana'],
                 'subdivision':['Ambala','Panchkula','Chandigarh','Mohali','Zirakpur','Kharar','Derabassi','Rajpura','Patiala','Ludhiana'],
                 'No_of_Units':[120,80,150,200,90,70,100,110,130,140],
                 'load':[1200, 1500, 800, 600, 900, 750, 1100, 1350, 1450, 1650]})
df
```

```python
# Increase size of Data
df_large = pd.concat([df]*10, ignore_index=True)
np.random.seed(42)
df_large['No_of_Units'] = df_large['No_of_Units'] + np.random.randint(-20, 20, size=len(df_large))
df_large['load'] = df_large['load'] + np.random.randint(-200, 200, size=len(df_large))  
df_large
```

```python
#concatination of two dataframes
df1=pd.DataFrame({'Circles':['Jalandhar','Hoshiarpur','Gurdaspur','Amritsar','Batala'],
                 'Division':['Jalandhar','Hoshiarpur','Gurdaspur','Amritsar','Amritsar'],
                 'subdivision':['Jalandhar','Hoshiarpur','Gurdaspur','  Amritsar','Batala'],
                 'No_of_Units':[125,85,155,205,95],
                    'load':[1250, 1550, 850, 650, 950]})
df_large=pd.concat([df_large,df1],ignore_index=True)
df_large
```

```python
#Adding Noise to data
np.random.seed(42)
df_large['No_of_Units'] = df_large['No_of_Units'] + np.random.randint(-20, 20, size=len(df_large))
df_large['load'] = df_large['load'] + np.random.randint(-200, 200, size=len(df_large))  
df_large    
```

```python
#add duplicates
df_large=df_large.append(df_large.iloc[0:5],ignore_index=True)
df_large
```

```python
#handling duplicates
df_large=df_large.drop_duplicates() 
```

```python
#handling missing values
df_large.loc[3,'No_of_Units']=np.nan
df_large.loc[7,'load']=np.nan
df_large['No_of_Units'].fillna(df_large['No_of_Units'].mean(),inplace=True)
df_large['load'].fillna(df_large['load'].mean(),inplace=True)
df_large
```

```python
#Data validation
def validate_data(df):
    assert df['No_of_Units'].min() >= 0, "No_of_Units contains negative values"
    assert df['load'].min() >= 0, "load contains negative values"
    assert df['Circles'].isnull().sum() == 0, "Circles column contains null values"
    assert df['Division'].isnull().sum() == 0, "Division column contains null values"
    assert df['subdivision'].isnull().sum() == 0, "subdivision column contains null values"
    print("Data validation passed.")
validate_data(df_large)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns   
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_large, x='No_of_Units', y='load', hue='Division')
plt.title('No_of_Units vs Load by Division')
plt.xlabel('No_of_Units')
plt.ylabel('Load')
plt.legend(title='Division')
plt.show()
```
