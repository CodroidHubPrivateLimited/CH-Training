```python
import requests
```

```python
url="https://dog.ceo/api/breeds/image/random"
```

```python
response = requests.get(url)
response
#response.status_code
```

```python
if response.status_code == 200:
    data = response.json()
    dog_image_url = data['message']
    print("here is the random dog image for you")
    print(dog_image_url)
```
