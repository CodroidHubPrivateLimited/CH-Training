```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(45)
n=100
data =({
    "ID":np.arange(1,n+1),
    "Name":np.random.choice(["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah"],n),
    "Age":np.random.randint(18,60,n),
    "Score":np.random.uniform(0,100,n),
    })
```

```python
df=pd.DataFrame(data)
df
```

```python
np.random.seed(45)
n=100
id=np.arange(1,n+1,5)
data =({
    "ID":id,
    "Name":np.random.choice(["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah"],len(id)),
    "Age":np.random.randint(18,60,len(id)),
    "Score":np.random.uniform(0,100,len(id)),
    })
```

```python
df=pd.DataFrame(data)
df
```
