```python
import cv2
import numpy as np
import sys
import requests
```

```python
# va= https://drive.google.com/file/d/1m2FbTzSsXd3MAb9rEr_BNMN_o0KwqTNu/view?usp=drive_link
file_id="1m2FbTzSsXd3MAb9rEr_BNMN_o0KwqTNu"
url=f"https://drive.google.com/uc?id={file_id}"

response=requests.get(url)
image_array=np.asarray(bytearray(response.content),dtype=np.uint8)
img_1=cv2.imdecode(image_array,-1)
print(img_1)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_1)
plt.axis('off')
plt.show()
```
