**Classes Concept in Python**

```python
class student:    #Class is created
  def __init__(self,name,marks):   #self is telling python on which data it is working
    self.name = name
    self.marks = marks
  def display(self):
    print(f"Name:{self.name} ,Marks:{self.marks}")
```

```python
stud1 = student("Amit",28)  #stud1 is an object of class student
```

**Industry level class creation**

```python
class Student:    #Class name should start with Caps
  def __init__(self, name:str ,marks:int):   #self is telling python on which data it is working
    self.name = name
    self.marks = marks

  def display(self)->str:  #add->string
    #print(f"Name:{self.name} ,Marks:{self.marks}") #instead use return keyword
    return f"Name:{self.name} ,Marks:{self.marks}"
```

```python
stud1 = Student("Amit",28)
print(stud1.display())
```

```python
stud2 = Student(42,28)
print(stud2.display())
type(stud2)
#def __init__(name: str, marks: int) it tells name should be string but not mandatorily.
```

```python
class Student:    #Class name should start with Caps
  def __init__(self, name:str ,marks:int):   #self is telling python on which data it is working
    if not isinstance(name,str):
      raise TypeError("Name should be string")
    if not isinstance(marks,int):
      raise TypeError("Marks should be integer")
    self.name = name
    self.marks = marks

  def display(self)->str:  #add->string
    #print(f"Name:{self.name} ,Marks:{self.marks}") #instead use return keyword
    return f"Name:{self.name} ,Marks:{self.marks}"
```

```python
stud2 = Student(42,28)
print(stud2.display())
```
