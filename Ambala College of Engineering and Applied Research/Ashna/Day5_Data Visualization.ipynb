**Step 3 Basic Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
df
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(tips_df['total_bill'],tips_df['tip']) #Passing x & y axis variables
plt.xlabel('Total Bill')    #Labelling for X-Axis
plt.title('Scatter plot of Total Bill vs Tip')   # Graph Heading
plt.ylabel('Tip($)')  #Labelling for Y-Axis
plt.show()  #To display the graph
```

```python
plt.figure(figsize=(8,6))
plt.pie(tips_df['total_bill']) #Passing x & y axis variables
plt.title('Pie Chart of Total Bill')   # Graph Heading
plt.show()  #To display the graph
```

```python
plt.figure(figsize=(8,6))
plt.boxplot(tips_df['total_bill']) #Passing x & y axis variables
plt.title('Box Plot of Total Bill')   # Graph Heading
plt.show()  #To display the graph
```
