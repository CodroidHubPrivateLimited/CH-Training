```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n=1000

df= pd.DataFrame({
    "cust_name" : np.random.choice(["Amit","Ajay", "Ankur","Reena", "Divya"]),
    "cust_age" : np.random.randint(18,60,size = n),
    "gender" : np.random.choice(["male","female"],size=n),
    "location" : np.random.choice(["ambala","Yamunanagar", "kurukshtra", "karnal"],size=n),
    "service usage": np.random.choice(["data","voice","all"],size=n),
    "tenure_info" : np.random.randint(0,10,size=n),
    "tenure_type" : np.random.choice(["monthly", "yearly"],size=n),
    "payment_behaviour" : np.random.choice(["on_time", "delayed"],size=n),
    #"service_quality" : np.random.choice(["complainent", "not_complainent"],size=n),
    "service_exp" : np.random.choice(["satisfactory","complainent"],size=n),
    "usage_in_hrs" :  np.random.randint(1,24,size=n),

})
df["monthly_charges"] = df["usage_in_hrs"]*12.87
df["total_charges"] = df["monthly_charges"]+ (0.18*df["monthly_charges"])
df["customer_churn"] = np.where((df["total_charges"]>100)&(df["service_exp"]=="complainent"),"yes","no")
```

```python
df
```

```python
bootstrap = df.sample(n=200, replace=True, random_state=42)
df1 = pd.concat([df, bootstrap], ignore_index=True)
df1
```

```python
valid_location = ["ambala","karnal","Yamunanagar","kurukshetra"]
df1 = df1['location'].isin(valid_location)
```
