### Lemmatization vs stemming

```python
import nltk
nltk.download("punkt") #for toknization
nltk.download("wordnet") #Dictionary used for lemmatization
nltk.download("omw-1.4") #helpswordnet to understand multiple word formats

from nltk.tokenize import word_tokenize
```

```python
text = "Deepti was explainig concepts very clearly. Some students complained that the lecture was boring and confusing."
text
```

```python
tokens = word_tokenize(text.lower())
tokens
```

```python
print(tokens)
```

```python
#stemming
from nltk.stem import PorterStemmer #popular rule-based stemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer #Dictonary based lemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos = 'v') for word in tokens]
print(lemmatized_words)
```

```python
print(tokens)
```

```python
print(stemmed_words)
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in lemmatized_words if word.lower() not in stop_words]
print(filtered_words)
```

```python
remove_punctuations = [word for word in filtered_words if word.isalpha()]
```

```python
remove_punctuations
```
