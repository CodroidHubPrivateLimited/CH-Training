```python
np.random.seed(42)
n=500
```

```python
data = pd.DataFrame({
    "attendance_rate": np.random.normal(75,10,n).clip(40,100),
    "avg_marks": np.random.normal(80,40,n),
    "assignments_submitted": np.random.normal(80,15,n).clip(20,100),
    "disciplinary_cases": np.random.poisson(1,n)
})
```

```python
data["dropout"] = (
    (data["attendance_rate"] < 60) |
    (data["avg_marks"] < 50) |
    (data["disciplinary_cases"] > 3)
)
```
