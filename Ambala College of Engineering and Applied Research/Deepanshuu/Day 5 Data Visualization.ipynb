```python
# count passenger by class

class_count=df['pclass'].value_counts()
print(class_count)
```

```python
class_count.plot(kind='bar', title='Number of Passengers by Class', xlabel='Passenger Class', ylabel='Number of Passengers')
```

```python
class_count.plot(kind='bar', title='age')
```
