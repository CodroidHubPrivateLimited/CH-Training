```python
import nltk

nltk.download('punkt') # usef for tokenization
nltk.download('wordnet') # Dictionary used for lemmatization
nltk.download('omw-1.4') # helps wordnet to understand multiple word formats

from nltk.tokenize import word_tokenize
```

```python
text = "Devashish was explaining concepts very clearly. Some students complained that the lecture was boring and confusing"
```

```python
text
```

```python
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
from nltk.stem import PorterStemmer # Popular rule-based stemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
```

```python
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer # Dictionary based lemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
```

```python
print(lemmatized_words)
```

```python
print(tokens)
```

```python
print(stemmed_words)
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
clean_words = [word for word in lemmatized_words if word not in stop_words]
```

```python
print(clean_words)
```

```python
clean_words1 = [word for word in clean_words if word.isalpha()]
print(clean_words1)
```
