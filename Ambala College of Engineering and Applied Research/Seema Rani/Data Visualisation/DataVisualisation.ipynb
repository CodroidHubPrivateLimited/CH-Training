```python
# Count of passengers by class
class_counts = df['pclass'].value_counts()
print(class_counts)
```

```python
#plot bar graph
class_counts.plot(kind='bar', title='Passenger Count by Class on Titanic', xlabel='Passenger Class', ylabel='Count')
```

```python
class_counts = df['age'].value_counts()
print(class_counts)
```

```python
class_counts.plot(kind='bar', title='Passenger Count by Class on Titanic', xlabel='Passenger Class', ylabel='Count')
```

```python
import pandas as pd
import seaborn as sns   
import matplotlib.pyplot as plt 

print(dir(sns))

#sns.set_theme(style="whitegrid")
```

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(tips_df['total_bill'], tips_df['tip'], color='blue')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill($)')
plt.ylabel('Tip($)')
plt.show()
```
