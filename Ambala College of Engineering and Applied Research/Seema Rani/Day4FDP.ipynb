```python
# Question: can we create a CSV file using python?
import csv  #used for creating csv files
header = ['Name', 'Age', 'City']
rows = [
    ['Amit', 32, 'Ambala'],
        ['Seema',39,'Ambala'],
        ['Deepti',36,'Delhi']
        ]


```

```python
header
```

```python
rows
```

```python
# create a CSV file
with open('friends.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)
print("CSV Created")    
```

```python
import sqlite3 as sql
connection=sql.connect('demo.db')
cursor=connection.cursor()
cursor.execute('''create table students( id integer Primary Key,
                      name text not null,
                      age integer,
                      city text);''')

print("Table created successfully")
connection.commit()
```
