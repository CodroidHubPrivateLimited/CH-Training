```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the Flipkart page
url = "https://www.flipkart.com/search?q=iphone"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all product elements
products = soup.find_all('div', class_='_1AtVbE')

# Extract product information
for product in products:
    name = product.find('div', class_='_4rR01T').text
    price = product.find('div', class_='_30jeq3 _1_WHN1').text
    print(f"Name: {name}, Price: {price}")
```
