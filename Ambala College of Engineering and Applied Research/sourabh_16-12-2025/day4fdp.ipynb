```python
#question: can we create csv file using python?
import csv # for csv files

header =['Name','age','city']

rows=[
    ['Amit',32,'Ambala'],
    ['Dev',31,'Patna'],
    ['Sonal',28,'Dehradun']
    ]
```

```python
print(header)
```

```python
print(rows)
```

```python
#create a csv file
with open('friends.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)

print("CSV Created")
```
