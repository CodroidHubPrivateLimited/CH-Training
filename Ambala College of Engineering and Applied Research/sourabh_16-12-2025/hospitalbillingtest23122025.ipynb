```python
import pandas as pd
import numpy as np
def register_patient():
    print("---- Patient Registration ----")

    patient_id = input("Enter Patient ID: ").strip()

    # Validate Patient Name
    while True:
        patient_name = input("Enter Patient Name: ").strip()
        if patient_name:
            break
        else:
            print("Error: Patient name cannot be empty.")

    # Validate Age
    while True:
        try:
            age = int(input("Enter Age: "))
            if age > 0:
                break
            else:
                print("Error: Age must be greater than 0.")
        except ValueError:
            print("Error: Please enter a valid integer for age.")

    # Validate Gender
    valid_genders = ["Male", "Female", "Other"]
    while True:
        gender = input("Enter Gender (Male/Female/Other): ").capitalize()
        if gender in valid_genders:
            break
        else:
            print("Error: Invalid gender. Choose Male, Female, or Other.")

    patient_record = {
        "Patient ID": patient_id,
        "Name": patient_name,
        "Age": age,
        "Gender": gender
    }

    print("\nPatient Registered Successfully!")
    print(patient_record)

    return patient_record


# Call function
register_patient()

    
```

```python
# Task 2: Capture Doctor Details

def capture_doctor_details():
    try:
        doctor_id = input("Enter Doctor ID: ").strip()
        
        doctor_name = input("Enter Doctor Name: ").strip()
        if doctor_name == "":
            raise ValueError("Doctor name cannot be empty.")

        specialization = input("Enter Specialization: ").strip()
        if specialization == "":
            raise ValueError("Specialization cannot be empty.")

        consultation_fee = float(input("Enter Consultation Fee: "))
        if consultation_fee <= 0:
            raise ValueError("Consultation fee must be greater than 0.")

        # Validated Doctor Record
        doctor_record = {
            "Doctor ID": doctor_id,
            "Doctor Name": doctor_name,
            "Specialization": specialization,
            "Consultation Fee": consultation_fee
        }

        print("\n Doctor record validated successfully!")
        print("Doctor Details:")
        for key, value in doctor_record.items():
            print(f"{key}: {value}")

    except ValueError as ve:
        print(f"\nValidation Error: {ve}")
    except Exception as e:
        print(f"\nUnexpected Error: {e}")


# Main Execution
capture_doctor_details()
```

```python
# Task 3: Appointment Booking

from datetime import datetime

# Sample doctor master data (normally comes from Task 2)
doctors = [
    {"Doctor ID": "D101", "Doctor Name": "Dr. Amit Sharma", "Specialization": "Cardiology"},
    {"Doctor ID": "D102", "Doctor Name": "Dr. Neha Verma", "Specialization": "Neurology"},
    {"Doctor ID": "D103", "Doctor Name": "Dr. Rajesh Gupta", "Specialization": "Orthopedics"}
]


def book_appointment():
    try:
        print("\nAvailable Doctors:")
        for doc in doctors:
            print(f"{doc['Doctor ID']} - {doc['Doctor Name']} ({doc['Specialization']})")

        doctor_id = input("\nSelect Doctor ID: ").strip()

        selected_doctor = None
        for doc in doctors:
            if doc["Doctor ID"] == doctor_id:
                selected_doctor = doc
                break

        if selected_doctor is None:
            raise ValueError("Invalid Doctor ID selected.")

        appointment_date_input = input("Enter Appointment Date (YYYY-MM-DD): ")
        appointment_date = datetime.strptime(appointment_date_input, "%Y-%m-%d").date()

        today = datetime.today().date()
        if appointment_date < today:
            raise ValueError("Appointment date cannot be in the past.")

        appointment = {
            "Doctor Name": selected_doctor["Doctor Name"],
            "Specialization": selected_doctor["Specialization"],
            "Appointment Date": appointment_date
        }

        print("\n Appointment Confirmed!")
        print("Appointment Details:")
        for key, value in appointment.items():
            print(f"{key}: {value}")

    except ValueError as ve:
        print(f"\n Validation Error: {ve}")
    except Exception as e:
        print(f"\n Unexpected Error: {e}")


# Main Execution
book_appointment()
```
