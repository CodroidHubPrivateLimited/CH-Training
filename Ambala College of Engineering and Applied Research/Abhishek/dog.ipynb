```python
import requests
```

```python
url="https://dog.ceo/api/breeds/image/random"
```

```python
response=requests.get(url)
```

```python
response.status_code
```

```python
if response.status_code == 200:
    data = response.json()
    dog_image_url = data['message']
    print("random image")
    print(dog_image_url)
else:
    print("Failed to retrieve data")
```
