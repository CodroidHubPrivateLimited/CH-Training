```python
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.tokenize import word_tokenize 
```

```python
text = "students attended the lecture. But they complained that it was boring. " 
text
 
```

```python
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
from nltk.stem import PorterStemmer


stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
```

```python
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word,pos='v') for word in tokens]
print(lemmatized_words)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_word = [ word for word in lemmatized_words if word not in stop_words]
print(filtered_word)
```

```python
clean_word1 = [word for word in filtered_word if word.isalpha()]
print(clean_word1)
```
