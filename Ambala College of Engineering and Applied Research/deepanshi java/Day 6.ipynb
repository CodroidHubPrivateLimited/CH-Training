```python
%pip install matplotlib 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
data ={
    "Age": np.random.randint(18, 60, 100),
    "Study_Hours": np.random.randint(1, 10, 100),
    "Attendance": np.random.randint(0, 100, 100),
    "Internal_Marks": np.random.randint(0, 50, 100),
    "Final _Marks": np.random.randint(0, 100, 100)
    
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Total_Marks"] = df["Internal_Marks"] + df["Final _Marks" ]
```

```python
df
```

```python
df["Placed"]= np.where((df["Total_Marks"]>=100) & (df["Attendance"]>75), "Yes", "No")
df
```

```python
df.groupby("Placed") ["Total_Marks"].mean()
```

```python
plt.figure()
plt.scatter(df["Study_Hours"], df["Total_Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Total Marks")
plt.title("Study Hours vs Total Marks")
```

```python
counts = df["Placed"].value_counts()
plt.figure()
counts.plot(kind='bar')
plt.xlabel("Eligible for Placement")
plt.ylabel("Number of Students")
plt.title("Placement Eligibility Distribution")
plt.show()
```
