```python
np.random.seed(20)
np.random.rand(3,100) 
employee_random={ 
    "emp_id":np.random.randint(100,200,100),
    "emp_name":np.random.choice(['Abhishek','Kumar','Singh','Ranjan','Ankit','Vikash'],100),
    "basic_month_sal":np.random.randint(20000,50000,100),
}

df_random=pd.DataFrame(employee_random)
print(df_random)
```
