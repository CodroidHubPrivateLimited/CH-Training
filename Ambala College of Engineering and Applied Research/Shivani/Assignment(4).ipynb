Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Adding Headers

```python
headers = {
"User-Agent": "Mozilla/5.0"
}
```

Requesting url

```python
url = "https://finance.yahoo.com/markets/mutualfunds/top/?start=0&count=25"
```

```python

response = requests.get(url, headers=headers)
```

Accessing the Code

```python
soup = BeautifulSoup(response.text, "html.parser")
```

Finding Table and its Rows that Contains the Details

```python
table = soup.find("table")
if table is None:
	print("Table not found. The data may be loaded dynamically")
else:
	tbody = table.find("tbody")
	if tbody is None:
		print("No tbody found in the table.")
		rows = []
	else:
		rows = tbody.find_all("tr")
```

Creating Empty list

```python
data = []
```

Accessing Required Information

```python
for row in rows:
	cols = row.find_all("td")
	data.append([col.get_text(strip=True) for col in cols])
```

```python
data
```

```python
print(data[0])
print(len(data[0]))
```

Naming Columns

```python
columns = [
"Symbol",
"Name",
"Last Price",
"Change",
"% Change",
"50 Days Average",
"200 Days Average ",
"3 Months Average",
"YTD return",
"52 Weeks Change%",
""
]
```

Converting into dataframe

```python
df = pd.DataFrame(data, columns=columns)
df
```

Saving into csv files

```python
df.to_csv("Top 25 Mutual Funds.csv")
```
