```python
import sqlite3 as sql


```

```python
connection=sql.connect('data.db')
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS student (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER
);
''')
```

```python
connection.commit()
```

```python
cursor.execute("insert into student (name, age) values ('Shivani', 22)")
```

```python
cursor.execute("select* from student")
results = cursor.fetchall()
for row in results:
    print(row)
```

```python
import pandas as pd
df=pd.DataFrame(results)
df
```
