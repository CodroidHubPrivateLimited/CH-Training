Gainer Crypto Currency

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
"User-Agent": "Mozilla/5.0"
}
```

```python
url = "https://finance.yahoo.com/markets/crypto/gainers/"
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table")
if table is None:
	print("Table not found. The data may be loaded dynamically")
	rows = []
else:
	tbody = table.find("tbody")
	if tbody is None:
		rows = []
	else:
		rows = tbody.find_all("tr")
```

```python
data = []
```

```python
for row in rows:
    cols = row.find_all("td")
data.append([col.get_text(strip=True) for col in cols])
data
print(data[0])
print(len(data[0]))
```

```python
 columns=[
"Symbol",
"Name",
" ",
"Price",
"Change",
"% Change",
"Market Cap",
"Volume ",
"Volume in currency( 24 hr)",
"Volume in all currencies(in 24 hr)",
"Circulating Supply",
"52 Weeks Change%",
" "
]
```

```python
df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
df.to_csv("Top Gainer Crypto currency.csv")
```
