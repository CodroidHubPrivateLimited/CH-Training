```python
#Stamming Vs Lemmatization
import nltk
nltk.download ('punkt') # used for tokenization
nltk.download ('wordnet') #Dictionaly for lemmatization
nltk.download ('omw-1.4') # helps wordnet to understand  multiple word formats
from nltk.tokenize import word_tokenize 
```

```python
text ="Priyanka was explaining  concepts very clearly.Some students complained that the  the lecture was boering and confusing"
```

```python
text
```

```python
tokens =word_tokenize(text.lower())
print(tokens)
```

```python
from nltk.stem import PorterStemmer # Popular Rule based Stemmer
stemmer =PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens] #convert into root word
print( stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer #Dictionary based lemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer_words=[lemmatizer.lemmatize(word,pos='v') for word in tokens] 
print (lemmatizer_words)#convert into root word
print (tokens)
```

```python
print (stemmed_words)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
clean_words = [word for word in lemmatizer_words if word not in stop_words]
print(clean_words)
```

```python
clean_words1 = [word for word in clean_words if word.isalpha()] #Remove Punctuation
clean_words1
```
