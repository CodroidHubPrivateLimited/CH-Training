```python
import pandas as pd
iris=sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species', s=100)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species', s=100)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species', s=100)

plt.title(' plot')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
#histplot
plt.figure(figsize=(8,6))
sns.histplot(data=iris, x='sepal_length',  kde=True, bins=20)#kds=kernal density estimation
plt.title(' plot')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=iris, x='species', y='petal_length', )

plt.title(' plot')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.pairplot(iris, hue='species') #Hue used for Different color Indication

plt.title(' plot')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```
