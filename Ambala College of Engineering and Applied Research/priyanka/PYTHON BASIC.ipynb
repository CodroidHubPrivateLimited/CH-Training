```python
class Student:
    def __init__(self,name:str,marks :int): #magic mthod is init
        self.name=name   #inbulit functions constructor call ing implicity
        self.marks=marks
    def display(self)->str:
        print (f"Name:{self.name}, Marks:{self.marks}")
        return (f"Name:{self.name}, Marks:{self.marks}")
stud2= Student("43",93)
print (stud2.display())
```

```python
class Student:
    def __init__(self,name:str,marks :int):
        if not isinstance(name,str):
            raise TypeError("name must be string" )
        if not isinstance(marks,int):
            raise TypeError("marks must be integer")
        self.name=name   #inbulit functions constructor call ing implicity
        self.marks=marks
    def display(self)->str:
        print (f"Name:{self.name}, Marks:{self.marks}")
        return (f"Name:{self.name}, Marks:{self.marks}")
stude2= Student("43",93)
print (stud2.display())
```
