Lemmatization Vs Stemming

```python
import nltk
```

```python
nltk.download('punkt') # use for tokenization
```

```python
nltk.download('stopwords')
nltk.download('wordnet') # Dictionary used for Lemmatization
nltk.download('omw-1.4') # helps wordnet to understand multiple word formats

from nltk.tokenize import word_tokenize
```

```python
text ="""Gagan was explaining concepts very clearly. 
Some students complained that the lecture was boring and confusing"""
```

```python
text
```

```python
tokens = word_tokenize(text.lower())
print(tokens)
```

```python
from nltk.stem import PorterStemmer # Popular rule-nased stemmer
```

```python
stemmer=PorterStemmer() 
```

```python
stemmed_words=[stemmer.stem(word) for word in tokens]
```

```python
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer # Dictionary based lemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
lemmatized_words =[lemmatizer.lemmatize(word, pos='v') for word in tokens]
```

```python
print(lemmatized_words)
```

```python
print(tokens)
```

```python
print(stemmed_words)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_stemmed_words = [word for word in stemmed_words if word.lower() not in stop_words]
print(filtered_stemmed_words)
```

```python
clean_words=[word for word in lemmatized_words if word not in stop_words]
```

```python
print(clean_words)
```

```python
clean_words1= [word for word in clean_words if word.isalpha()]
```
