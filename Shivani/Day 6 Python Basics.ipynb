```python
class student:
    def __init__(self, name: str, marks: int):
        self.name = name
        self.marks = marks
    
    def display(self) -> str:
        print(f"Name: {self.name}, Marks: {self.marks}")
        return f"Name: {self.name}, Marks: {self.marks}"
```

```python
stud1 = student("Dev", "85")
stud1.display()
```

# Students Performance and Placement

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data ={
    "Age": np.random.randint(18, 60, 100),
    "Study_hours": np.random.randint(1, 10, 100),
    "Attendance": np.random.randint(50, 100, 100),
    "Internal_marks": np.random.randint(40, 90, 100),
    "Final_marks": np.random.randint(60, 95, 100)
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Total_marks"] = df["Internal_marks"] + df["Final_marks"]
```

```python
df
```

```python
df["Placed"] =np.where(df["Total_marks"] >= 75, "Yes", "No"
    )
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
df["Placed"].value_counts()
```

```python
df.groupby("Placed").mean()
```

```python
plt.figure()
plt.scatter(df["Study_hours"], df["Final_marks"])
plt.xlabel("Study Hours")
plt.ylabel("Final Marks")
plt.title("Study Hours vs Final Marks")
```

```python
plt.figure()
df["Placed"].value_counts().plot(kind='bar')
plt.xlabel("Eligibility")  
plt.ylabel("Number of Students")
plt.title("Placement Eligibility Distribution")  
```
