Capture Employee Details (Input Validation)

```python
import pandas as pd
import numpy as np
```

```python
# =====================================================
# Task 1: Capture Employee Details (Input Validation)
# =====================================================

emp_id = input("Enter Employee ID: ")

name = input("Enter Employee Name: ")
if not name.isalpha():
    raise ValueError("Employee name must contain only alphabets")

department = input("Enter Department: ")

total_working_days = int(input("Enter Total Working Days in Month: "))
if total_working_days <= 0:
    raise ValueError("Total working days must be greater than 0")


# =====================================================
# Task 2: Capture Attendance Data
# =====================================================

days_present = int(input("Enter Days Present: "))
if days_present < 0 or days_present > total_working_days:
    raise ValueError("Invalid attendance data")


# =====================================================
# Task 3: Attendance Percentage Calculation
# =====================================================

attendance_percentage = (days_present / total_working_days) * 100


# =====================================================
# Task 4: Attendance Status Classification
# =====================================================

if attendance_percentage >= 90:
    attendance_status = "Excellent"
elif attendance_percentage >= 75:
    attendance_status = "Satisfactory"
elif attendance_percentage >= 60:
    attendance_status = "Needs Improvement"
else:
    attendance_status = "Poor"


# =====================================================
# Task 5: Leave Balance Initialization
# =====================================================

leave_balance = 2


# =====================================================
# Task 6: Leave Request Validation
# =====================================================

leave_requested = int(input("Enter Leave Days Requested: "))

if leave_requested <= leave_balance:
    leave_approved = leave_requested
    leave_status = "Approved"
else:
    leave_approved = 0
    leave_status = "Rejected"


# =====================================================
# Task 7: Leave Deduction Logic
# =====================================================

leave_balance = leave_balance - leave_approved


# =====================================================
# Task 8: Appraisal Eligibility Check
# =====================================================

if attendance_percentage >= 75 and leave_balance >= 0:
    appraisal_eligible = "Yes"
else:
    appraisal_eligible = "No"


# =====================================================
# Task 9: Monthly Attendance Summary (Procedural)
# =====================================================

print("\n--- Monthly Attendance Summary ---")
print("Employee ID       :", emp_id)
print("Employee Name     :", name)
print("Attendance %      :", round(attendance_percentage, 2))
print("Attendance Status :", attendance_status)
print("Leave Balance     :", leave_balance)
print("Appraisal Eligible:", appraisal_eligible)


# =====================================================
# Task 10: Store Attendance Records
# =====================================================

attendance_records = []

attendance_records.append({
    "Employee ID": emp_id,
    "Name": name,
    "Department": department,
    "Attendance %": round(attendance_percentage, 2),
    "Status": attendance_status,
    "Leave Balance": leave_balance,
    "Appraisal Eligible": appraisal_eligible
})


# =====================================================
# Task 11: EmployeeAttendance Class Design (OOP)
# =====================================================

class EmployeeAttendance:
    def __init__(self, emp_id, name, department, total_days, days_present):
        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.total_days = total_days
        self.days_present = days_present
        self.leave_balance = 2


# =====================================================
# Task 12: Attendance Calculation Method
# =====================================================

    def calculate_attendance_percentage(self):
        return (self.days_present / self.total_days) * 100


# =====================================================
# Task 13: Leave Processing Method
# =====================================================

    def process_leave_request(self, leave_requested):
        if leave_requested <= self.leave_balance:
            self.leave_balance -= leave_requested
            return "Approved"
        return "Rejected"


# =====================================================
# Task 14: Appraisal Eligibility Method
# =====================================================

    def check_appraisal_eligibility(self):
        return self.calculate_attendance_percentage() >= 75 and self.leave_balance >= 0


# =====================================================
# Task 15: Final Attendance Report Generation
# =====================================================

employee = EmployeeAttendance(
    emp_id, name, department, total_working_days, days_present
)

print("\n====== Final Attendance Report ======")
print("Employee ID        :", employee.emp_id)
print("Name               :", employee.name)
print("Department         :", employee.department)
print("Attendance %       :", round(employee.calculate_attendance_percentage(), 2))
print("Attendance Status  :", attendance_status)
print("Leave Balance      :", employee.leave_balance)
print("Appraisal Eligible :", 
      "Yes" if employee.check_appraisal_eligibility() else "No")
print("====================================")
```
