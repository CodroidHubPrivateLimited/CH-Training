```python
import seaborn as sns
df=sns.load_dataset('titanic')
```

```python
#Count of passengers based by class
class_counts = df['class'].value_counts()
print(class_counts)
```

```python
class_counts.plot(kind='bar', title='Number of Passengers by Class')
```

```python
class_counts = df['fare'].value_counts()
print(class_counts)
class_counts.plot(kind='bar', title='Number of Passengers by Fare')
```

```python
class_counts = df['age'].value_counts()
print(class_counts)
class_counts.plot(kind='bar', title='Number of Passengers by Age')
```

```python
class_counts = df['class'].value_counts()
print(class_counts)
class_counts.plot(kind='pie', title='Number of Passengers by Class')
```

```python
class_counts = df['class'].value_counts()
print(class_counts)
class_counts.plot(kind='box', title='Number of Passengers by Class')
```

```python
class_counts = df['class'].value_counts()
print(class_counts)
class_counts.plot(kind='hist', title='Number of Passengers by Class')
```
