```python
books=soup.find_all('div',class_='book')
```

```python
print (f"Number of books found: {len(books)}")
```

```python
book1_data = []
for book in books:
    title = book.find('p', class_='title').text.strip()
    author = book.find('p', class_='author').text.strip()
    price = book.find('p', class_='price').text.strip()
    publication_year = book.find('p', class_='publication-year').text.strip()
    genre = book.find('p', class_='genre').text.strip()

    description = book.find('p', class_='description').text.strip()

    book1_data.append ( {
    'Title' : title,
    'Author' : author,
    'Price': price,
    'Publication Year': publication_year,
    'Genre': genre,
    'Description': description,

})

    df=pd.DataFrame(book1_data)
    print(df)
```
