```python
import nltk
nltk.download('punkt')  #for tokenization
nltk.download('wordnet') #dictionary used for lemmatization
nltk.download('stopwords') #removes extra words
nltk.download('omw-1.4') #helps wordnet to understand multiple word formats
from nltk.tokenize import word_tokenize
```

```python
text ="Ashna was explaining concepts very clearly ! Some students complained that the lecture was boring and confusing"
```

```python
tokens = word_tokenize(text.lower()) #Text is tokenized
tokens  #vertical words split
```

```python
print (tokens) #one line
```

```python
from nltk.stem import PorterStemmer #popular rule-based stemmer
stemmer = PorterStemmer()
stemmer_words = [stemmer.stem(word) for word in tokens]
stemmer_words
```

```python
from nltk.stem import WordNetLemmatizer #Dictionary based lemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word,pos='v') for word in tokens]
lemmatized_words
```

```python
from nltk.corpus import stopwords #Removing stop words
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
clean_words = [word for word in lemmatized_words if word not in stop_words]
clean_words
```

```python
clean_words1 = [word for word in clean_words if word.isalpha()] #Removing punctuation marks
clean_words1
```

```python
nltk.download('averaged_perceptron_tagger_eng')   #For POS Required
```

```python
nltk.pos_tag(clean_words1)
```
