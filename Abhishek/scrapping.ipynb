```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
url= "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=4178b939-1f74-48ff-b88a-f2b095776502&as-backfill=on"
```

```python
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
if soup.body:
    soup = soup.body
```

```python
for tag in  soup(["script", "style", "head", "noscript", "meta", "link"]):
    tag.decompose()
```

```python
print(soup.prettify())
```

```python
product=soup.find_all('div',class_='jIjQ8S')
```

```python
product
```

```python
print(len(product))
```

```python
Laptop = []

for i in product:
    title = i.find('div', class_='_4rR01T')
    price = i.find('div', class_='_30jeq3 _1_WHN1')
    
    Laptop.append({
        "Name": title.get_text(strip=True) if title else None,
        "Price": price.get_text(strip=True) if price else None,
    })

print(Laptop)
```
