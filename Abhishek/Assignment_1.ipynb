```python
np.random.seed(1)
n = 20

ids = np.arange(1, n+1, 5)
data = {
    "ID": ids,
    "Name": np.random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'], len(ids)),
    "Age": np.random.randint(18, 65, len(ids)),
    "score": np.random.randint(0, 100, len(ids))
}
df = pd.DataFrame(data)
print(df)
```
