```python
import csv
header = ['Name', 'Age', 'City']
rows = [['amit',32,'ambala'], ['sumit',28,'delhi'], ['rahul',30,'bangalore']] 
header
```

```python
rows
```

```python
with open('friends.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)

print("CSV file created successfully.")
```

```python
import sqlite3 as sql
```

```python
connection = sql.connect('demo.db')
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("create table students (id integer PRIMARY KEY, name text not null, age integer)");
```

```python
connection.commit()
```

```python
cursor.execute("insert into students (id, name, age) values (1, 'Amit', 32)");
```

```python
cursor.execute("select * from students")
results = cursor.fetchall()
for row in results:
    print(row)
```

```python
import pandas as pd
df = pd.DataFrame(results)
df
```
