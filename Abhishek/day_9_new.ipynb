```python
token = word_tokenize(text)
print(token)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_token = [word for word in token if word.lower() not in stop_words]
print(filtered_token)
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer= PorterStemmer()
stemmed_words=[stemmer.stem(word) for word in token]
```

```python
print("Original Text:", token)
```

```python
print("Stemmed Text:",stemmed_words)
```

```python
def stemmedtext(text):
    token=word_tokenize(text)
    from nltk.stem import PorterStemmer
    stemmer= PorterStemmer()
    return [stemmer.stem(word) for word in token]
```

```python
text2="The children were running quickly, they will be playing until evening"
stemmewords = stemmedtext(text2)
print(stemmewords)
```
