```python
#question : can we created csv file using python?
import csv # for csv file handling
header = ['Name', 'Age', 'City'] # header for csv file (list)
rows = [['Deepanshi', 22, 'Delhi'], # data for csv file (list of lists)
        ['Aman', 24, 'Mumbai'],
        ['Riya', 21, 'Bangalore']]
header
rows
```

```python
#create a CSV file
with open('friends.csv', mode = 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # write the header
    writer.writerows(rows) 
    
```

```python
print("csv created") 
```
