```python
import numpy as np
import pandas as pd

```

```python
np.random.seed(0)
n = 500
df = pd.DataFrame({
    "Age": np.random.randint(18, 60, size=n),
    "CGPA": np.random.randint(0, 120, size=n),
    "Internships_Count": np.random.randint(0, 10, size=n),
    "Technical_Score": np.random.randint(0, 120, size=n),
    "Attendance_Percentage": np.random.randint(50, 101, size=n),
})
df
```

```python
#Target variable with imbalance
df['Placed'] = np.where((df['CGPA'] > 7.5) & (df["Technical_Score"] > 70), "yes", "no")
```

```python
bootstrap = df.sample(n=200, replace=True, random_state=42)
df_aug = pd.concat([df, bootstrap], ignore_index=True)
```

```python
df_aug
```

```python
noise = df.sample(n=150, random_state=42).copy()
noise["CGPA"] += np.random.normal (0, 0.1, 150)
noise["Technical_Score"] += np.random.normal (-3, 4, 150)
noise["Attendance_Percentage"] += np.random.normal (0, 2, 150)
```

```python
# log constraints
noise["CGPA"] = noise["CGPA"].clip(6.0, 9.5)
noise["Technical_Score"] = noise["Technical_Score"].clip(50, 100)
noise["Attendance_Percentage"] = noise["Attendance_Percentage"].clip(70, 100)
```

```python
df_aug = pd.concat([df_aug, noise], ignore_index=True)
df_aug
```
