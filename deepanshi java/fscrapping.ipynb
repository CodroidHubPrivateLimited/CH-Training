```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'}
URL = "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=" \
"laptop%7CLaptops&requestId=85d42fba-4f4d-414b-8479-5589b4628f43&as-searchtext=laptp"
response = requests.get(URL, headers=HEADERS)
response
soup=BeautifulSoup(response.content, 'html.parser')
soup
print(soup.prettify())


```

```python
if soup.body:
    soup = soup.body
for tag in soup(["script", "style","head","noscript", "meta", "link"]):
    tag.decompose()
print (soup.prettify())
```

```python
products = soup.find_all('div', class_='jIjQ8S')
print(len(products))
```

```python
products
```

```python
laptops = []
for product in products:
    title = product.find('div', class_='RG5S1K')
    price = product.find('div', class_='hZ3P6w DeU9vF')
    laptops.append({'Title': title.get_text(strip=True) if title else None,
'Price': price.get_text(strip=True) if price else None})
```

```python
df = pd.DataFrame(laptops)
df

```
