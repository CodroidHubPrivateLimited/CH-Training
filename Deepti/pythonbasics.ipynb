```python
class student:
    def __init__(self, name, marks): # work as constructor (magic method)
        self.name = name
        self.marks = marks
    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")
        
```

```python
student1 = student("Deepti", 95)
```

```python
student1.display()
```

```python
student2 = student(88, "Alex")
```

```python
student2.display()
```

```python
class Student:
    def __init__(self, name: str, marks: int): # work as constructor (magic method)
        self.name = name
        self.marks = marks
    def display(self)->str:
        return(f"Name: {self.name}, Marks: {self.marks}")
        
```

```python
stud1 = Student(45, 95)
print(stud1.display())
```

```python
class Student:
    def __init__(self, name: str, marks: int): # work as constructor (magic method)
        if not isinstance(name, str):
            raise TypeError("Name must be a string")
        if not isinstance(marks,int):
            raise TypeError
        self.name = name
        self.marks = marks
    def display(self)->str:
        return(f"Name: {self.name}, Marks: {self.marks}")
```

```python
stud1 = Student(45, "amit")
print(stud1.display())
```
