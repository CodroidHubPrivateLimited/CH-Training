```python
class Customer:
    def __init__(self, cust_id: int, cust_name: str, delivery_city: str ): # work as constructor (magic method)
        if not isinstance(cust_id, int):
            raise TypeError("ID must be a interger")
        if not isinstance(cust_name, str):
            raise TypeError(" Customer Name must be a string")
        #if isinstance(delivery_city==Null):
            raise ValueError(" Delivery city must not be empty")
        self.cust_id = cust_id
        self.cust_name = cust_name
        self.delivery_city = delivery_city
    def display(self)->str:
        return(f"Cutomer_ID: {self.cust_id}, Customer_Name: {self.cust_name}, Delivery_City: {self.delivery_city}")
```

```python
cust1 = Customer(2193, "Amit", "Mumbai")
print(cust1.display())
```
