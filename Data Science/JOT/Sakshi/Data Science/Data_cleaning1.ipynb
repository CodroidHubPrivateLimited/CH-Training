## **Identification:**  Use df.isnull().sum() to count missing values per column.


```python
df.isnull().sum()  # Check missing values
```

## **Deletion:** Remove rows or columns with a high percentage of missing values using dropna().

```python
df.dropna()  # Remove rows with missing data
```

## **Imputation:** Fill missing values using methods like fillna() with a specified value (e.g., mean, median, mode) or a forward/backward fill.

```python
df.fillna(0)      # Replace missing values with 0
```
