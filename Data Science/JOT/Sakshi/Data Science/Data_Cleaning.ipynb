```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.amazon.in/s?k=tops+for+women+western&i=apparel&crid=E7U7JPHM57AK&sprefix=tops+%2Capparel%2C260&ref=nb_sb_ss_mvt-t11-ranker_1_5"
```

```python
response=requests.get(url)
response
```

```python
headers={    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
response=requests.get(url,headers=headers)
response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
soup= BeautifulSoup(response.content,"html.parser")
#soup
```

```python
# print(soup.prettify())
```

```python
#tops=soup.find_all("div",class_="sg-col-inner")
#tops
```

```python
tops=soup.find_all("div",class_="a-section a-spacing-base")
#tops
```

```python
len(tops)
```

```python
print(len(tops))
```

```python
top1=[]
```

```python
for top in tops:
    top_name=top.find("span",class_="a-size-base-plus a-color-base")
    if top_name:
        Title=top_name.get_text(strip=True)
    Total_price=top.find("span",class_="a-price-whole")
    if Total_price:
        Price=Total_price.get_text(strip=True)
    top1.append(
        {'Title':Title,
    'Price':Price,
        }
    )
```

```python
print(top1)
```

```python
df=pd.DataFrame(top1)
```
