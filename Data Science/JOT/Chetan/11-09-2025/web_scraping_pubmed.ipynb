```python
from bs4 import BeautifulSoup as b 
import requests as r
import pandas as pd
```

```python
url='https://pubmed.ncbi.nlm.nih.gov/?term=paracetamol&filter=datesearch.y_1&filter=years.2024-2026'
headers={    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/536.37 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response=r.get(url,headers=headers)
response
```

```python
med=b(response.content,'html.parser')
```

```python
print(med.prettify())
```

```python
all_med=med.find_all('article',class_='full-docsum')
```

```python
len(all_med)
```

```python
med_data=[]
```

```python
for meds in all_med:
    title=meds.find('a',class_='docsum-title')
    descr=meds.find('div',class_='full-view-snippet')
    i_d=meds.find('span',class_='docsum-pmid')
    if title and descr and i_d:
        title=title.get_text(strip=True)
        descr=descr.get_text(strip=True)
        i_d=i_d.get_text(strip=True)
    med_data.append(
        {
            'Titile':title,
            'Description':descr,
            'ID':i_d
        }
    )
```

```python
d=pd.DataFrame(med_data)
```
