## 01. Write a function to check whether a given number is Armstrong (e.g., 153 → 1³+5³+3³ = 153). 

```python
def armst():
    number = input("Enter a number: ")
    digits = []

    for i in number:
        digits.append(int(i))

    length = len(digits)
    total = 0

    for digit in digits:
        total += digit ** length

    if total == int(number):
        print(number, "is an Armstrong number.")
    else:
        print(number, "is not an Armstrong number.")


armst()


```
