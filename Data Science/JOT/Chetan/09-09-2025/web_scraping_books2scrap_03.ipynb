```python
from bs4 import BeautifulSoup as b 
import requests as r
import pandas as pd
```

```python
url='https://books.toscrape.com/catalogue/category/books/poetry_23/index.html'
response=r.get(url)
```

```python
response
```

```python
books=b(response.content,'html.parser')
books
```

```python
# all_books
print(books.prettify())
```

```python
all_books=books.find_all('li',class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
len(all_books)
```

```python
books_data=[]
```

```python
for book in all_books:
    # isd=1
    name=book.find("h3").find("a")["title"]
    price=book.find('p',class_="price_color").get_text(strip=True)

    if price and name :
        books_data.append(
        {
            'Name':name,
            'Price':price
        }
        )
```

```python
books_data
```

```python
d=pd.DataFrame(books_data)
```
