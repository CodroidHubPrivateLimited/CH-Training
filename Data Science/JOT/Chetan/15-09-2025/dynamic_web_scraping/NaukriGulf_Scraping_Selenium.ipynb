```python
#Every line pulls in a different piece.

#This gives you the main interface for controlling a browser. webdriver.Chrome() will literally launch Chrome that Selenium can steer.
from selenium import webdriver 

#A Service object tells Selenium where the ChromeDriver executable is located and how to run it. 
#When you use ChromeDriverManager, it plugs right into this.
from selenium.webdriver.chrome.service import Service

#Lets you configure Chrome’s behavior: run in headless mode (no window), disable notifications, set custom user-agents, etc.
from selenium.webdriver.chrome.options import Options

#Provides different ways of locating elements on the page: by ID, by CSS selector, by XPath, etc.
from selenium.webdriver.common.by import By

#Lets you pause until something you care about is loaded, instead of scraping too early.
from selenium.webdriver.support.ui import WebDriverWait

#Works with WebDriverWait to define conditions like “wait until this element is visible” or “wait until this element is clickable.”
from selenium.webdriver.support import expected_conditions as EC

#The automation butler we installed earlier. This fetches the right version of ChromeDriver for your system and plugs it into Selenium.
from webdriver_manager.chrome import ChromeDriverManager   #ChromeDriver for Chrome, GeckoDriver for Firefox''

#Finally, BeautifulSoup is your HTML parser. 
#Once Selenium renders the page, you hand the HTML source over to BeautifulSoup to dig out the pieces you actually want.
from bs4 import BeautifulSoup
```

```python
"""
    Selenium = the driver + waiter,
    webdriver-manager = the driver-fetcher,
    BeautifulSoup = the plate and cutlery you use to actually consume the data.
"""
None
```

```python
chrome_options = Options()                         #Creates a configuration object for Chrome.
chrome_options.add_argument("--headless")  
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/131.0.6778.265 Safari/537.36"
)
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
# driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
```

```python
url = "https://www.naukrigulf.com/top-jobs-by-designation"
driver.get(url)
```

```python
WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CLASS_NAME, "soft-link"))
    )
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
job_profiles_section = soup.find_all('a', class_='soft-link darker')

print("Top Job Profiles:")
for i, job in enumerate(job_profiles_section[:10], start=1):
    print(f"{i}. {job.text.strip()}")
```

```python
driver.quit()
```
