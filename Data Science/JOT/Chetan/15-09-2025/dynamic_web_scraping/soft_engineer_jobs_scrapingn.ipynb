```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
```

```python
chrome_options = Options()                         #Creates a configuration object for Chrome.
chrome_options.add_argument("--headless")
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/131.0.6778.265 Safari/537.36"
)
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
url = "https://www.foundit.in/search/software-engineer-jobs"
driver.get(url)
```

```python
job_titles = WebDriverWait(driver, 30).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "text-darkKnight-700"))
)
print(f"Found {len(job_titles)} job titles")
# print
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
job_profiles=soup.find_all('h3',"text-darkKnight-700")

for i,job in enumerate(job_profiles[:25],start=1):
    print(f'{i}. {job.text.strip()}')
```

```python
import pandas as pd
df=pd.DataFrame(job_profiles, columns=["top Job Title"])
```
