```python
import requests
url = "https://api.open-meteo.com/v1/forecast"
```

```python
params = {
    "latitude": 52.52,
    "longitude": 13,
}

response = requests.get(url, params=params)
response.raise_for_status()
data1 = response.json()
print(data1)
```

```python
params = {
    "latitude": 35,
    "longitude": 12
}

response = requests.get(url, params=params)
response.raise_for_status()
data2 = response.json()
print(data2)
```
