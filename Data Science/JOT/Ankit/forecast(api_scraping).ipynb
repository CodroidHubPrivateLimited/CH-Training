```python
import requests

# Step 1: Fetch raw data from CoinGecko API
url = "https://api.open-meteo.com/v1/forecast"
params = {"latitude": 52.52,
          "longitude":13.41,
          "hourly":"temperature_2m"
         }

response = requests.get(url, params=params)
response.raise_for_status()

# Raw JSON data
data0 = response.json()
print(data0) 
```
