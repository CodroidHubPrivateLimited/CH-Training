```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs 
```

```python
url="https://www.amazon.in/s?k=mouse&crid=2FIR9LHRJNUKO&sprefix=mouse%2Caps%2C275&ref=nb_sb_noss_2"
response=requests.get(url)
response
```

```python
headers={    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/536.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
response=requests.get(url,headers=headers)
response
```

```python
if response.status_code==403:
    print("error(oops...)")
elif response.status_code==200:
    print("Wonderful....")
```

```python
soup=bs(response.content,'html.parser')
soup
```

```python
print(soup.prettify)
```

```python
mouse=soup.find_all('div',class_="a-section a-spacing-small a-spacing-top-small")
mouse
```

```python
len(mouse)
```

```python

mo=[]
for i in mouse:
    print(i)
    name=i.find('h2',class_="a-size-medium a-spacing-none a-color-base a-text-normal").find('span').get_text(strip=True)
    price=i.find('span',class_="a-offscreen").get_text(strip=True)
    rating=i.find('span',class_="a-icon-alt").get_text(strip=True)
    dl=i.find('div',class_="a-row a-color-base udm-primary-delivery-message")
    
    mo.append({
        'price': price,
        'name': name,
        'rating': rating,
        'delivery':dl
        
        
    })
print("done")
    
```

```python
df=pd.DataFrame(mo)
df
```
