<h2><p>Data Cleaning  is the process of identifying and fixing or removing errors and inconsistencies in data to improve its quality. This is a critical step in the data analysis pipeline to ensure that your analysis or model is built on accurate, complete, and reliable data. Data Cleaning is also called <b>data cleansing</b> or <b>scrubbing</b>



<div style="background-color: lightgreen; color: black; padding: 4px;">
<h1>-->Why is Data Cleaning Important?
</h1></div>
<h2><p><ul><li>
Improves<b> accuracy</b> of analysis and models
</li><li>
 Ensures <b>valid results</b> from data insights
</li><li> Eliminates<b> duplicate, corrupted, or incomplete data</b>
</li><li>
 Helps maintain <b> data consistency</b> across sources
</li></ul>



<div style="background-color: lightgreen; color: black; padding: 4px;">
<h1>--> Data Cleaning Tasks
</h1></div>

<h2>
<p><b> 1. Handling Missing Data:-</b>Missing data is a common problem in datasets. Strategies to handle missing data include:
<ul>
<li><u>Removing Records:</u> Deleting rows with missing values if they are relatively few and insignificant.
</li>
<li><u>Imputing Values:</u> Replacing missing values with estimated ones, such as the mean, median, or mode of the dataset.
</li>

</ul></p>

<p><b> 2. Removing Duplicates:-</b>Duplicates can skew analyses and lead to inaccurate results. Identifying and removing duplicate records ensures that each data point is unique and accurately represented.</p>

<p><b>3. Correcting Inaccuracies:-</b>Data entry errors, such as typos or incorrect values, need to be identified and corrected. This can involve cross-referencing with other data sources or using validation rules to ensure data accuracy.</p>
</p>

<p><b>4.  Standardizing Data:-</b>Standardization means bringing data to a common format, like:
<ul>
<li><u>Date formats</u> (2025-07-29 vs 29-07-2025)
</li>
<li><u>Units</u> (kg, lbs)
</li>
<li><u>
Capitalization</u> (INDIA, India, india → India)
</li></ul>

</p>







</h2>
 


<div style="background-color: lightgreen; color: black; padding: 4px;">
<h1>-->Challenges in Data Cleaning
</h1></div>
<h2><p><ul><li><b>Volume of Data:</b> Large datasets can be challenging to clean due to their sheer size. Efficient techniques and tools are necessary to handle big data cleaning tasks.</li></p>
<p><li><b>
Complexity of Data:</b> Data from diverse sources may have different structures and formats, making it difficult to clean and integrate.</li></p><p><li><b>
Continuous Process:</b> Data cleaning is not a one-time task but an ongoing process. As new data is collected, it needs to be continually cleaned and maintained.
</li></ul></p></h2>

<h1><p><b>Handle these errors or anomalies</b></p></h1>
<h2><p><ul><li><u><i>Listwise deletion:-</i></u> removes entire rows from the dataset if any single value is missing or erroneous in that row.</li>
<li><u><i> Pairwise Deletion:-</i></u>Pairwise deletion keeps all available data for each analysis, only excluding missing values when necessary.
</li>
</ul></p></h2>

```python
 # drop rows with any missing values
df_listwise_deletion = df.dropna()
```

```python
# drop rows with missing values in 'Ticket' and 'Age' columns
df_pairwise_deletion = df.dropna(subset=['Fare']) 
```
