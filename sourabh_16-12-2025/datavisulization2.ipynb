```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(8,6))
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=iris, x='sepal_length', hue='species', multiple='stack')
plt.title('plot of sepal length')
plt.xlabel('sepal length')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=iris, x='species', y='sepal_length', palette='Set2')
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('species')
plt.ylabel('sepal length')
plt.show()
```

```python
sns.pairplot(iris, hue='species', palette='Set1')
plt.show()
```
