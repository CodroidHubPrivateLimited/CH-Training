Modeling is the process of converting a real time problem into a mathematical function that can learn patterns from data and predictions on unseen data.
Modeling Pipeline
1. Problem Definition
    Goal: Predict whether a student will drop[1] or continue[0]
    Type: Supervised learning --> Binary Classification
2. Feature Selection
   Features:
   1. Attendance
   2. Average Marks
   3. Assignments Submitted
   4. Disciplinary cases
3. Test-Train Split
    80%--Traning
    20%--Testing
4. Algorithm Selection
    No algorithm performs best in all scenarious/for all problems
    Model Comparisons
    1. Accuracy
    2. Recall
    3. Interoperability
    4. Stability
5. Model Training
6. Evaluation and Comparison
    Use Multiple Metics
    1. Confusion 
        Actual/Predicted            No Dropout[0]           Dropout[1]
        No Dropout[0]                     TN                      FP
        Dropout[1]                        FN                      TP
    
                  
    2. Core Evaluation Matrices
        2.1 Accuracy        TP+TN/TP+TN+FP+FN   Overall Correctness
        2.2 Precision       TP/TP+FP            How reliable dropout predictions
        2.3 Recall          TP/TF+FN            How mqny actual deopouts we catchec
        2.4 F1-Score        2*Precision* Recall/Precision+Recall. it is balance betwenn Precision and Recall
        2.5 ROC-AUC         it is a curve



7. Final Model Selection

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


```

```python
from sklearn.metrics import(confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score)
```

```python
np.random.seed(42)
n=500
```

```python
data = pd.DataFrame({
    "attendance_rate":np.random.normal(75,10,n).clip(40,100),
    "avg_marks":np.random.normal(80,40,n).clip(30,100),
    "assignments_submitted":np.random.normal(80,15,n).clip(20,100),
    "disciplinary_cases":np.random.poisson(1.2,n)
})
```

```python
data["dropout"] =(
    (data["attendance_rate"]<60) |
    (data["avg_marks"]<50) |
    (data["disciplinary_cases"]>3)
).astype(int)
```

```python
data
```

```python
X = data.drop("dropout", axis=1)
```

```python
print(X)
```

```python
y = data["dropout"]
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
print(X_train_scaled)
```

```python
models = {
    "Logistic Regression":LogisticRegression(max_iter=1000),
    "Decision Tree":DecisionTreeClassifier(max_depth=5),
    "Random Forest":RandomForestClassifier(n_estimators=100),
    "SVM":SVC(probability=True)
}
```

```python
result=[]

for name, model in models.items():
    if name in["Logistic Regression","SVM"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_prob = model.predict_proba(X_test_scaled)[:1]
    else:
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:1]
```

```python
cm =  confusion_matrix(y_test,y_pred)
```

```python
result.append({
    "Model": name,
    "Accuracy":accuracy_score(y_test,y_pred),
    "Precision":precision_score(y_test,y_pred),
    "Recall":recall_score(y_test,y_pred),
    "FI-Score":f1_score(y_test,y_pred),
    "ROC-AUC":roc_auc_score(y_test,y_pred),
    "Confusion Matrix":cm
})
```

```python
result_df = pd.DataFrame(result)
print(result_df)
```

```python
for r in result:
    print(f"\n{r['Model']}Confusion Matrix:")
    print(r["Confusion Matrix"])
```
