Modeling is the process of converting a real time problem into a mathematical function that can learn patterns from data and predictions on unseen data.
Modeling Pipeline
1. Problem Definition
    Goal: Predict whether a student will drop[1] or continue[0]
    Type: Supervised learning --> Binary Classification
2. Feature Selection
   Features:
   1. Attendance
   2. Average Marks
   3. Assignments Submitted
   4. Disciplinary cases
3. Test-Train Split
    80%--Traning
    20%--Testing
4. Algorithm Selection
    No algorithm performs best in all scenarious/for all problems
    Model Comparisons
    1. Accuracy
    2. Recall
    3. Interoperability
    4. Stability
5. Model Training
6. Evaluation and Comparison
    Use Multiple Metics
    1. Confusion 
        Actual/Predicted            No Dropout[0]           Dropout[1]
        No Dropout[0]                     TN                      FP
        Dropout[1]                        FN                      TP
    
                  
    2. Core Evaluation Matrices
        2.1 Accuracy        TP+TN/TP+TN+FP+FN   Overall Correctness
        2.2 Precision       TP/TP+FP            How reliable dropout predictions
        2.3 Recall          TP/TF+FN            How mqny actual deopouts we catchec
        2.4 F1-Score        2*Precision* Recall/Precision+Recall. it is balance betwenn Precision and Recall
        2.5 ROC-AUC         it is a curve
