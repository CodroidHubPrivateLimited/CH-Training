```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#responding url

url='https://www.amazon.in/s?k=facewash+men&crid=2PQKEQQERU5C5&sprefix=facewash%2Caps%2C255&ref=nb_sb_ss_saint-en-refocus-candidate_ci_hl-bn-left_1_8'

response=requests.get(url)
```

```python
if response.status_code == 200:
     print("Successfully fetched the page!")
else:
     print("Failed to retrieve the page. Status code:{response.status_code}")
```

```python
#Accessing the Code

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#prititfy the code
print(soup.prettify())
```

```python
#Finding All Division Class That Contains the Details

products=soup.find_all("div",class_="sg-col-inner")
```

```python
#finding the number of division class
print(len(products))
```

```python
products
```

```python
#creating empty list 
Facewash=[]
```

```python
#accessing required information

for product in products:
    title = price = None
    title_elem = product.h2
    if title_elem:
     title = title_elem.get_text(strip=True)
    price_elem=product.find("span",class_="a-offscreen")
    if price_elem:
     price=price_elem.get_text(strip=True)
    if title and price:
     Facewash.append({
    'Title': title,
    "Price": price,
})
```

```python
print(Facewash)
```

```python
#converting into dataframe
df=pd.DataFrame(Facewash)
```

```python
df
```

```python
#removing irrelevant values
df= df[df['Title'].str.strip() != '']
irrelevant_keywords = ['results for', 'Results', 'Trending', 'Related', 'Need␣help',"Today's deal"]
pattern = '|'.join(irrelevant_keywords)
df = df[~df['Title'].str.contains(pattern, case=False, na=False)]
```

```python
df = df.reset_index(drop=True)
```

```python
#saving to CSV file
df.to_csv("Amazon Facewash Details.csv")
```
