##NLTK=  Natural language Toolkit   excalidraw 

it is a python library that provides ready to use liguistic resources and algorithms to process and analysis human language

how NLTK works?
1. Raw text :  NLTK is AMAZING!!! It helps students understand NLP easily.
2. text processing : to clean and standardize text so that later stages work correctly.
                     Operations - lower caseing, remove punctuations, remove special characters, remove numbers
                     nltk is amazing it helps students to understand nlp easily
3. Tokenization:    break text into smaller units(tokens) that NLP algorithm cn work with it.
                    Types
                   1. sentence tokenization
                   2. word tokenization
                   ['nltk', 'is', 'amazing', 'it', 'helps' ,'students' 'to' ,'understand', 'nlp' ,'easily']
4. normalization
                    1. Removing STOP words:
                    words like is, it , the carry little meaning in most NLP tasks- STOP WORDS
                    ['nltk', 'amazing', 'helps' ,'students' ,'understand', 'nlp' ,'easily']

                    2. Limitization(meaningful root words):
                    to reduce words to their dictionary form

                    
5. linguistic analysis
                        This add language intelligence
                        1. POS Tagging: Identifies grammtical role of each word
                        [('nltk','NN') ('amazing,'JJ'), ('helps',vB) ,('students','NN') ,('understand',VB), ('nlp','NN') ,('easily','RB')]

                        2. NER- Named Entity Recongition(optional)
                        Identifies entities like-person, organization, location

6. structured output
                     Text is now converted into structured, machine-usable data

                     a. Token-POS Tagging
                      [('nltk','NN') ('amazing,'JJ'), ('helps',vB) ,('students','NN') ,('understand',VB), ('nlp','NN') ,('easily','RB')]

                      b. Feature Vectors
                      [''nltk':1'amazing':1,....]
