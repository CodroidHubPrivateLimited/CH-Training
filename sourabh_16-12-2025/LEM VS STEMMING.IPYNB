```python
import nltk
nltk.download('wordnet')               #Dictionary used for lemmitization
nltk.download('omw-1.4')          #helps wordnet to udnerstand multiple languages
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
text="Sourabh is learning NLTK library. He has been learning it since last year. He learned about stemming and lemmatization."
```

```python
text
```

```python
tokens=word_tokenize(text.lower())
print(tokens)
```

```python
tokens
```

```python
from nltk.stem import PorterStemmer    #this is a rule based stemmer
stemmer=PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in tokens]
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer    #this is a dictionary based lemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
```

```python
print(lemmatized_words)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
word_to_remove='since'
filtered_tokens = [word for word in filtered_tokens if word != word_to_remove]
print(filtered_tokens)
```
