Capture Customer & Product Details

```python
class Customer:
    def __init__(self, Customer_name,Producer_name,Price,Quantity): # work as constructor (magic method)
        self.Customer_name=Customer_name
        self.Producer_name= Producer_name
        self.Price=Price
        self.Quantity=Quantity

    def display(self):
        print(f"Customer_name: {self.Customer_name}, Producer_name: {self.Producer_name},Price:{self.Price}, Quantity:{self.Quantity}")
```

```python

customer1 = Customer("Gagan", "Kevin",500,12)
```

```python
customer1.display()
```
