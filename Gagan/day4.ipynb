```python
import sqlite3 as sql
```

```python
connection=sql.connect('demo.db')
cursor=connection.cursor()
```

```python
cursor.execute('''
    Create table student (
    id INTEGER PRIMARY KEY, 
    name TEXT not null, 
    age INTEGER
);''')
```

```python
connection.commit()
print("Table created successfully.")
```

```python
cursor.execute("insert into student (id,name, age) values (1, 'Gagan', 22)")
cursor.execute("insert into student (id,name, age) values (2, 'Amit', 32)")
cursor.execute("insert into student (id,name, age) values (3, 'Dev', 31)")
```

```python
cursor.execute("select * from student;")
results=cursor.fetchall()
for row in results:
    print(row)
```

```python
connection.commit()
print("Table created successfully.")
```
