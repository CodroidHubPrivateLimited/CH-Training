```python
np.random.seed(41)
n=4000
df=pd.DataFrame({
    "Patient_ID":np.arange(1,n+1),
    "first_name":np.random.choice(["John", "Jane", "Jim", "J    ill", "Jack"],n),
    "last_name":np.random.choice(["Doe", "Smith", "Brown", "Davis", "Wilson"],n),
    "gender":np.random.choice(["M", "F"],n),
    "date_of_birth":pd.to_datetime(np.random.randint(1950,2005,size=n),format="%Y")+pd.to_timedelta(np.random.randint(1,365,size=n),unit="D"),
    "Age":np.random.randint(18,80,size=n),
    "Blood_group":np.random.choice(["A+","A-","B+","B-","O+","O-","AB+","AB-"],n),
    "City":np.random.choice(["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],n),
    "State":np.random.choice(["NY", "CA", "IL", "TX", "AZ"],n),
    "address":np.random.choice(["123 Main St", "456 Oak St", "789 Pine St", "101 Maple St", "202 Birch St"],n),
    "country":np.random.choice(["USA", "Canada", "UK", "Australia", "India"],n),
    "medical_history":np.random.choice(["Hypertension",  "Diabetes",  "Asthma", "Allergies","Cancer"],n),
    "Registration_date":pd.to_datetime(np.random.randint(2015,2023,size=n),format="%Y")+pd.to_timedelta(np.random.randint(1,365,size=n),unit="D")   
})
```

```python
df
```

```python
for col in["Patient_ID","first_name","last_name","gender","date_of_birth","Age","Blood_group","Contact_number","City","State","address","country","medical_history","Registration_date" ]:
    df.loc[df.sample(frac=0.05).index, col] = np.nan  # Introduce 5% missing values
df 
```

```python
pd.concat([df, df.sample(20)], ignore_index=True) # Duplicate 20 random rows

#Introduce invalid ages
df.loc[df.sample(10).index, "Age"] = np.random.choice([-5, 240], size=10)  
df
```

```python
df.loc[df.sample(10).index, "Marks"] = np.random.choice([150, 300], size=10)
df
```

```python
df["medical_history"] = df["medical_history"].astype(str)
df.loc[df.sample(frac=0.1).index, "medical_history"] = df["medical_history"].sample(10).str.lower()
df
```
