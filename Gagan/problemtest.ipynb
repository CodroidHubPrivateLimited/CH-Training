```python
np.random.seed(41)
n=100
df=pd.DataFrame({
    "Amount":np.random.randint(5000,15000,n),
    "Age": np.random.randint(18, 24,size=n),
    "Income": np.random.randint(10000,40000,n),
    "Claim_Amount": np.random.randint(10000,31000,n),
    "Payment_delay": np.random.randint(1,5,n),
    "Policy": np.random.choice(["lifeinsurance","healthisurance"], size=n)
})
```

```python
df
```

```python
df.head()
```

```python
for col in["Policy","Amount","Age","Income","Claim_Amount","Payment_Delay"]:
    df.loc[df.sample(frac=0.05).index, col] = np.nan  # Introduce 5% missing values
df 
```

```python
#Introdece Duplicate records/Rows
pd.concat([df, df.sample(20)], ignore_index=True) # Duplicate 20 random rows

#Introduce invalid ages
df.loc[df.sample(10).index, "Age"] = np.random.choice([-5, 240], size=10)  
df
```

```python
#Noise Injection
noise=df.sample(n=100, random_state=42).copy()
noise['Age'] += np.random.normal(20,30,100)
noise['Income'] += np.random.normal(-3, 4,100)
noise['Amount'] += np.random.normal(0, 2,100)

#Logical Constraints
noise['Age'] = noise['Age'].clip(6.0,9.5)
noise['Income'] = noise['Income'].clip(55,100)
noise['Amount'] = noise['Amount'].clip(65,100)
df_aug=pd.concat([df_aug,noise], ignore_index=True)
df_aug
```
