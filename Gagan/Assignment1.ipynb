```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36␣ (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
url = 'https://www.amazon.in/s?k=facewash&crid=1JIWRT8KWWKOL&sprefix=%2Caps%2C361&ref=nb_sb_ss_recent_2_0_recent'
response = requests.get(url)
if response.status_code == 200:
  print("Successfully fetched the page!")
else:
  print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
products=soup.find_all("div",class_="sg-col-inner")

```

```python
print(len(products))
```

```python
products
```

```python
Facewash=[]
for product in products:
  title = price = None
  title_elem = product.h2
if title_elem:
  title = title_elem.get_text(strip=True)
  price_elem=product.find("span",class_="a-offscreen")
if price_elem:
 price=price_elem.get_text(strip=True)
if title and price:
 Facewash.append({
 'Title': title,
 "Price": price,
})
```

```python
print(Facewash)
```

```python
df=pd.DataFrame(Facewash)
df
```

```python
df= df[df['Title'].str.strip() != '']
irrelevant_keywords = ['results for', 'Results', 'Trending', 'Related', 'Need␣help',"Today's deal"]
pattern = '|'.join(irrelevant_keywords)
df = df[~df['Title'].str.contains(pattern, case=False, na=False)]
```

```python
df = df.reset_index(drop=True)
```
