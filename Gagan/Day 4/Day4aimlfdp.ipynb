```python
# Question: Can we createa csv file using python?
```

```python
import csv # for csv files
header = ['Name', 'Age', 'City']    
rows = [
    ['Amit', 32, 'Ambala'],
    ['Dev', 31, 'Patna'],
    ['Sonal', 3528, 'Dehradun']
       ]

header
```

```python
rows
```

```python
# Create a csv file
with open('friends.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # Write the header
    writer.writerows(rows)   # Write the data rows
print("CSV created successfully.")
```
