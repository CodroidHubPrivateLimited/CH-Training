```python
#Task10
class Product:
    def __init__(self, product_id, name, price, category):
        if price <= 0:
            raise ValueError("Price must be greater than 0")
        self.product_id = product_id
        self.name = name
        self.price = price
        self.category = category
```

```python
#task10-11
class Bill:
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.products = []   # list of tuples (product, quantity)
        self.subtotal = 0
        self.gst = 0
        self.discount = 0
```

```python
# Add product to bill
def add_product(self, product, quantity):
    if quantity < 1:
        raise ValueError("Quantity must be at least 1")
    self.products.append((product, quantity))
```

```python
# Task 12: GST Calculation
    def calculate_gst(self):
        gst_total = 0
        for product, qty in self.products:
            item_total = product.price * qty
            if product.category.lower() == "essentials":
                gst_total += item_total * 0.05
            elif product.category.lower() == "electronics":
                gst_total += item_total * 0.18
        self.gst = gst_total
        return gst_total
```

```python
 # Task 13: Discount Calculation
 # def apply_discount(self):
        if self.subtotal >= 10000:
            self.discount = self.subtotal * 0.10
        elif self.subtotal >= 5000:
            self.discount = self.subtotal * 0.05
        else:
            self.discount = 0
        return self.discount
```

```python
# Task 14 & 15: Invoice Generation
def generate_invoice(self):
    print("\n========== INVOICE ==========")
    print(f"Customer Name : {self.customer_name}")
    print("\nProducts:")
    for product, qty in self.products:
        print(f"{product.name} | ₹{product.price} x {qty} = ₹{product.price * qty}")

    print("------------------------------")
    print(f"Subtotal      : ₹{self.subtotal:.2f}")
    print(f"GST           : ₹{self.gst:.2f}")
    print(f"Discount      : ₹{self.discount:.2f}")
    print(f"Final Amount  : ₹{self.subtotal + self.gst - self.discount:.2f}")
    print("==============================\n")
```

```python
# =========================
# Main POS Workflow
# =========================
def main():
    customer_name = input("Enter Customer Name: ")
    bill = Bill(customer_name)

    while True:
        print("\nEnter Product Details")
        pid = input("Product ID: ")
        name = input("Product Name: ")
        price = float(input("Price per unit: "))
        qty = int(input("Quantity: "))
        category = input("Category (Essentials/Electronics): ")

        product = Product(pid, name, price, category)
        bill.add_product(product, qty)

        more = input("Add more products? (yes/no): ")
        if more.lower() != "yes":
            break
```
