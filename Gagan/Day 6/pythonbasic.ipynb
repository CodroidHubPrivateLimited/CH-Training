```python
class Student:   # industry level(classname capital and define type)
    def __init__(self,name: str,marks:int): #magic method
        if not isinstance(name,str):
            raise TypeError("name must be string")
        if not isinstance(marks, int):
            raise TypeError("marks should be integer")
        self.name = name
        self.marks =marks

    def display(self)->str:
        #print(f"Name: {self.name}, Marks: {self.marks} ")  
        return f"Name: {self.name},  Marks: {self.marks}" 
```

```python
stud2 =Student("dev",98)
print(stud2.display())
```

```python
stud3 =Student("dev","Gagan")
print(stud3.display())
```
