```python
import requests
url="https://dog.ceo/dog-api/"
url
```

```python
response=requests.get(url)
response
```

```python
response.json()
```

```python
if response.status_code==200:
    data=response.json()
    dog_image_url=data['message']
    print("Random Dog Image URL:", dog_image_url)
    print(dog_image_url)
else:
    print("Failed to retrieve data from the API")
```
