```python
url = 'https://www.amazon.in/s?k=facewash&crid=1JIWRT8KWWKOL&sprefix=%2Caps%2C361&ref=nb_sb_ss_recent_2_0_recent'
response = requests.get(url)
```

```python
if response.status_code==200:
    print("Successfully accessed the webpage")
else:
    print("Failed to retrieve the webpage")
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
products=soup.find_all('div',{'data-component-type':'s-search-result'})
```

```python
print(len(products))
```

```python
products
```

```python
Facewash=[]
```

```python
for product in products:
    title=price=None
    title_elem=product.h2
    if title_elem:
        title=title_elem.get_text(strip=True)
    price_elem=product.find("span",class_="a-offscreen")
    if price_elem:
        price=price_elem.get_text(strip=True)   
        if title and price:
            Facewash.append({
                'Title':title,
                'Price':price})
```

```python
print(Facewash)
```

```python
df=pd.DataFrame(Facewash)
```

```python
df
```

```python
import re

# ensure df is a DataFrame (in case it became a list)
if isinstance(df, list):
	df = pd.DataFrame(df)

# ensure 'Title' column exists
if 'Title' not in df.columns:
	df['Title'] = ''

# ensure Title values are strings to avoid list/string indexing issues
df['Title'] = df['Title'].astype(str)

# drop empty/whitespace titles
df = df[df['Title'].str.strip() != '']

irrelevant_keywords = ['results for', 'Results', 'Trending', 'Related', 'Needhelp', "Today's deal"]
pattern = '|'.join(re.escape(k) for k in irrelevant_keywords)
df = df[~df['Title'].str.contains(pattern, case=False, na=False)]
```

```python
df = df.reset_index(drop=True)
```
