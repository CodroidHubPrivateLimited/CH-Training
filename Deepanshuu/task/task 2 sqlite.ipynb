```python
import sqlite3 as sql
```

```python
connection=sql.connect('data.db')
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
create table student(
    id integer PRIMARY KEY,
    name text not null,
    age integer
);''')
```

```python
connection.commit()
```

```python
cursor.execute("select * from student")
results=cursor.fetchall()
for row in results:
    print(row)
```
