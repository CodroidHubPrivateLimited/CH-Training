```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('tips')   
```

```python
df.head()
```

```python
# count passenger by class

class_count=df['tip'].value_counts()
print(class_count)
```

```python
class_count.plot(kind='bar', title='tip')
```

```python
# count fair by class

class_count=df['day'].value_counts()
print(class_count)
```

```python
class_count.plot(kind='bar', title='day')
```

```python
class_count=df['time'].value_counts()
print(class_count)
```

```python
class_count.plot(kind='bar', title='time')
```
