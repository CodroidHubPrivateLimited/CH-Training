```python
import sqlite3 as sql
```

```python
connection=sql.connect('demo.db')
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
create table student(
    id integer PRIMARY KEY,
    name text not null,
    age integer
);''')
```

```python
connection.commit()
```

```python
cursor.execute("insert into student(id,name,age) values(1,'Deepanshu',23)")
```

```python
cursor.execute("select * from student")
results=cursor.fetchall()
for row in results:
    print(row)
```

```python
import pandas as pd
df=pd.DataFrame(results)
df
```

```python
connection.close()
```
