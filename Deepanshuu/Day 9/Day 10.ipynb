```python
import nltk
nltk.download('omw-1.4') # helps wordnet to understand multiple word formats

from nltk.tokenize import word_tokenize
```

```python

nltk.download('wordnet')# dict. used for lemmatization
```

```python
text = "Deepanshu was explaining concepts very clearly but unforyunately. Some students complainted that lecture was boring and confusing"
```

```python
text
```

```python
tokens = word_tokenize(text.lower())
```

```python
print(tokens)
```

```python
from nltk.stem import PorterStemmer # Popular rule-based stemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
```

```python
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_word = [lemmatizer.lemmatize(word, pos='v')for word in tokens]
```

```python
print(lemmatized_word)
```

```python
print(tokens)
```

```python
print(stemmed_words)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
lemmatized_word = [word for word in tokens if word.lower() not in stop_words]
print(lemmatized_word)
```

```python
filtered_word = [word for word in lemmatized_word if word not in stop_words]
print(filtered_word)
```

```python
# remove punctuations
clean_word1= [word for word in filtered_word if word.isalpha()]
print(clean_word1)
```
