```python
# Question: Can we create  a csv file using python?
import csv # lib. for csv

header = ['Name', 'Age', 'City'] # header row

rows=[
    ['Deepanshu', 23, 'Ambala'],
    ['Shivani', 24, 'Kurukshetra'],
    ['Abhishek', 26, 'Yamunanagar']
]


```

```python
header
```

```python
rows
```

```python
# Create a csv file
with open('friends.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # write header
    writer.writerows(rows)   # write data 
print("CSV file created successfully.")
```
