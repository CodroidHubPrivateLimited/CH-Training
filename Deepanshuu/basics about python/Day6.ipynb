#### STUDENT PERFORMANCE AND PLACEMENT

```python
%pip install matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
data ={
    "Age": np.random.randint(20,25,100),
    "Study_Hours": np.random.randint(1,8,100),
    "Attendance": np.random.randint(60,100,100),
    "Internal_Marks": np.random.randint(40,90,100),
    "Final_Marks": np.random.randint(60,95,100)

}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python

df["Total_Marks"] = df["Internal_Marks"] + df["Final_Marks"]
```

```python
df
```

```python
df["Placed"] = np.where(
    (df["Total_Marks"] >= 120) & (df["Attendance"] >= 75), "Yes", "No"
)
```

```python
df.groupby("Placed") ["Total_Marks"].mean()
```

```python
plt.figure()
plt.scatter(df["Study_Hours"], df["Final_Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Final Marks")
plt.title("Study Hours vs Final Marks")
```

```python
counts = df["Placed"].value_counts()
plt.figure()
counts.plot(kind='bar')
plt.xlabel("Eligible for Placement")
plt.ylabel("Number of Students")
plt.title("Placement Eligibility Distribution")
plt.show()

```
