```python
class Student:
    def __init__(self, name: str, marks: int):
        if not isinstance(name, str):
            raise TypeError("Name must be a string")
        if not isinstance(marks, int):
            raise TypeError("Marks must be an integer")
        self.name = name
        self.marks = marks
    
    def display(self)-> str:
        #print(f"Name: {self.name}, Marks: {self.marks}")
        return f"Name: {self.name}, Marks: {self.marks}"
        
```

```python
stud2 = Student("Deepanshu", 95)

stud2.display()
```

#### assert keyword in python - used for debugging purposes. It tests a condition as true. If the condition is false, it raises an AssertionError with an optional message.
#### Syntax: assert condition  |  assert condition, "Error message" 


```python
marks1 = -60
assert marks1 >= 0
```

```python
marks1 = -60
assert marks1 >= 0, "Marks cannot be negative"
```

```python
class student:
    def __init__(self, name, marks):
        assert isinstance(name, str), "Name must be a string"
        assert isinstance(marks, int), "Marks must be an integer"
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")
        
```

```python
s1 = student(23,45)
s1.display()
```
