```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n=2000
```

```python
#Creating Customer Data
Customer_details =({
    "customer_id" :np.random.randint(100,n+1),
    "customer_age" : np.random.randint(18,65,2000),
    "customer_gender" : np.random.choice(['M','F'],2000),
    "customer_income" : np.random.randint(20000,100000,2000),
    "delivery_time" : np.random.randint(10,30,2000),
    "customer_location" : np.random.choice(['Urban','Suburban','Rural'],2000),
    "customer_occupation" : np.random.choice(['Professional','Business','Student','Unemployed'],2000),
    "customer_channel" : np.random.choice(['organic','Ads','referral'],2000)

})
df= pd.DataFrame(Customer_details)
df
```

```python
df.info()
```

```python
df["customer_satisfaction"] = np.where(
    (df["delivery_time"] < 15) & (df["customer_location"] == "Urban"),"No","Yes")
df
```

```python
#Creating Synthetic data with Augmentation technique-bootstrap
bootstrap= df.sample(n=1000,replace=True)
syn_data=pd.concat([df,bootstrap],ignore_index=True)
syn_data
```
