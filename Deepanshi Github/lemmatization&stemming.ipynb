```python
#triple """ use for multiline 
text = """Deepanshi was explaining concepts very clearly.
Some students complained that the lecture was boring and confusing"""
text
```

```python
tokens = word_tokenize(text.lower())

print (tokens)
```

```python
from nltk.stem import PorterStemmer # Popular rule-based stemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
```

```python
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word,pos='v')for word in tokens]
```

```python
print(lemmatized_words)
```

```python
print(tokens)
```

```python
print(stemmed_words)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

# Get English stop words
stop_words = set(stopwords.words('english'))

# Remove stop words from tokens
filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens)
```

```python
clean_words = [word for word in lemmatized_words if word not in stop_words]
print(clean_words)
```

```python
import string

# Remove punctuation and stop words
clean_tokens = [word for word in tokens if word not in string.punctuation and word not in stop_words]

print(clean_tokens)
```
