```python
def bootstrap_ci(data, n_bootstraps=10000, ci=95):
    bootstrapped_means = []
    for _ in range(n_bootstraps):
        resampled_data = np.random.choice(data, size=len(data), replace=True)
        bootstrapped_means.append(np.mean(resampled_data))
    lower = np.percentile(bootstrapped_means, (100 - ci) / 2)
    upper = np.percentile(bootstrapped_means, (100 + ci) / 2)
    return np.mean(data), lower, upper
```

```python
def introduce_missing_values(df, columns, prob=0.05):
    for col in columns:
        mask = np.random.rand(len(df)) < prob
        df.loc[mask, col] = np.nan
    return df
```
