```python
import pandas as pd 
import numpy as np 
```

```python
np.random.seed(0)
n = 500
df = pd.DataFrame({
    "Student ID": np.arange(1, n+1),
    "Name": np.random(['Amit', 'Asha', 'Ankit', 'Parul', 'Rahul'], size=n),
    "Course Name": np.random.choice(['B.Tech', 'M.Tech', 'MBA', 'BBA', 'PhD'], size=n),
    "Subjects": np.random.choice(['Math', 'Science', 'History', 'Art', 'Physical Education'], size=n),
    "Marks": np.random.randint(0, 101, size=n),
})
df
```

```python

df['Marks'] = np.random.randint(0, 101, size=n)
```

```python

total_marks = df['Marks'].sum()
df
```

```python

percentage = (total_marks / (n * 100)) * 100
percentage
```

```python
gpa = (percentage / 100) * 4
gpa
```
