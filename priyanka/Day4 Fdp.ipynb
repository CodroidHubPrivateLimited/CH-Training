```python
import csv # for csv files
header =['Name','Age','city']
rows =[
    ['Amit',32,'Ambala'],
    ['deV',32,'PATNA'], ['sONAL',28, 'Dehradun']]
header
rows
```

```python
#create a csv file
with open ('friends.csv', mode='w' , newline='') as file:
    writer =csv.writer(file)
    writer.writerows(header)
    writer.writerows(rows)

    print ("csv created")
```
