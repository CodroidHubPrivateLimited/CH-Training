```python
# Data Augmentation
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
n=500
df = pd.DataFrame ({
    "Age": np.random.randint(20,26,n),
    "CGPA" : np.round(np.random.uniform(6.0,9.5,n), 2),
    "Internship_Count": np.random.randint(0,4,n),
    "Technical_Score" : np.random.randint(50,100, n),
    "Attendence_Percentage": np.random.randint (60,100,n)
    ""
    
})
df
```

```python
bootstrap = df.sample(n=200, replace=True, random_state=42)
df_aug = pd.concat([df, bootstrap], ignore_index=True)
df_aug
```

```python

#Noise Injection
noise = df.sample(n=150, random_state=42).copy()
noise['CGPA'] += np.random.normal(0, 0.1, 150)
noise['Technical_Score'] += np.random.normal(-3, 4, 150)
noise['Attendence_Percentage'] += np.random.normal(0, 2, 150)

#Logical Constraints
noise['CGPA'] = noise['CGPA'].clip(6.0, 9.5)
noise['Technical_Score'] = noise['Technical_Score'].clip(55, 100)
noise['Attendence_Percentage'] = noise['Attendence_Percentage'].clip(65, 100)
df_aug = pd.concat([df_aug, noise], ignore_index=True)
df_aug
```
