```python
import requests                 
from bs4 import BeautifulSoup    
import pandas as pd  
```

```python
url = 'https://www.goodreads.com/quotes'
response = requests.get(url)
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
 }
```

```python
 response = requests.get(url, headers=headers)
 response
```

```python
if response.status_code == 200:
    print("Successfully fetched the page!")
else:
    print(f"Failed to retrieve the page. Status code:{response.status_code}")
```

```python
soup1 = BeautifulSoup(response.content, 'html.parser')
```

```python
soup1
```

```python
print(soup1.prettify())
```

```python
quotes=soup1.find_all("div" ,class_='quote')
```

```python
quotes
```

```python
 print(f"Number of qoutes found: {len(quotes)}")
```

```python
quotes
```

```python
quotes_data=[]
```

```python
for i in quotes:
    title_div = i.find("div", class_="quoteText").get_text( strip=True)
    Author=i.find("span",class_="authorOrTitle").get_text(strip=True)

    quotes_data.append({
    "Quote":title_div,
    "Author":Author
    }) 
```

```python
quotes_data
```

```python
df=pd.DataFrame(quotes_data)
```

```python
df
```

```python
df.to_csv("Popular Quotes.csv")
```
