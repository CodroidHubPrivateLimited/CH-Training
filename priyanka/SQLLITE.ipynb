```python
import sqlite3 as sql
```

```python
connection = sql.connect('demo.db')
```

```python
cursor = connection.cursor()
cursor.execute('''CREATE TABLE deepu (id integer primary key, name text not null, age integer)''')
```

```python
connection.commit()
```

```python

cursor.execute("INSERT INTO deepu (id, name, age) VALUES (2, 'Deepanshi', 22)")
#connection.execute("")
```

```python
cursor.execute("SELECT * FROM deepu")
results = cursor.fetchall()
for row in results:
    print(row)
```

```python
import pandas as pd
df = pd.DataFrame (results)
df
```
