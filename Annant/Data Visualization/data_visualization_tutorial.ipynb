# Data Visualization

Welcome to this data visualization tutorial! In this notebook, we will explore various techniques for visualizing data using Python's popular libraries, `matplotlib` and `seaborn`. Data visualization is a crucial skill for understanding and communicating insights from data.

## 1. Setting up our Environment

First, we need to import the necessary libraries. We will use:
- `pandas` for data manipulation
- `matplotlib` for creating basic plots
- `seaborn` for more advanced and aesthetically pleasing plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set_theme(style="whitegrid")
```

## 2. Loading the Data

For this tutorial, we will use the `tips` dataset, which is built into the `seaborn` library. This dataset contains information about tips given in a restaurant.

```python
tips_df = sns.load_dataset('tips')

# Display the first 5 rows of the dataset
tips_df.head()
```

## 3. Basic Plots with Matplotlib

`matplotlib` is the foundational library for plotting in Python. Let's create some basic plots.

### Scatter Plot

A scatter plot is great for showing the relationship between two numerical variables.

```python
plt.figure(figsize=(8, 6))
plt.scatter(tips_df['total_bill'], tips_df['tip'])
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

### Bar Chart

A bar chart is useful for comparing a numerical variable across different categories.

```python
average_tip_by_day = tips_df.groupby('day')['tip'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(average_tip_by_day['day'], average_tip_by_day['tip'], color=['blue', 'green', 'red', 'purple'])
plt.title('Average Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip ($)')
plt.show()
```

### Histogram

A histogram is used to see the distribution of a single numerical variable.

```python
plt.figure(figsize=(8, 6))
plt.hist(tips_df['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

## 4. Enhanced Plots with Seaborn

`seaborn` is built on top of `matplotlib` and provides a high-level interface for drawing attractive and informative statistical graphics.

### Box Plot

A box plot shows the distribution of a numerical variable across different categories, highlighting the median, quartiles, and outliers.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips_df)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Violin Plot

A violin plot is a combination of a box plot and a kernel density plot. It shows the distribution of the data.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='day', y='total_bill', data=tips_df, inner='quartile')
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

### Pair Plot

A pair plot allows us to see the relationship between all pairs of numerical variables in a dataset. It's great for exploratory data analysis.

```python
sns.pairplot(tips_df, hue='sex', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables in the Tips Dataset', y=1.02)
plt.show()
```

### Heatmap

A heatmap is a graphical representation of data where values are depicted by color. It is useful for visualizing the correlation between variables.

```python
# Calculate the correlation matrix
correlation_matrix = tips_df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of the Tips Dataset')
plt.show()
```
